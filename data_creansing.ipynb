```python
import numpy as np
import pandas as pd

data = pd.read_csv('./data/tweet_activity_metrics_yassh_i_20200829_20200926_en.csv')

data.head()
```

```python
df = pd.DataFrame(
    {
        'Tweet text' : data['Tweet text'],
        'impressions' : data['impressions'],
        'time' : pd.to_datetime(data['time']) + pd.to_timedelta(9,unit='h'),
        'engagements' : data['engagements'],
        'engagement rate' : data['engagement rate'],
        'retweets' : data['retweets'],
        'replies' : data['replies'],
        'likes' : data['likes']
    }
)
df.set_index('time', inplace=True)
df.head()
```
